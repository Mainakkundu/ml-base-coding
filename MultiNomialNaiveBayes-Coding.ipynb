```python
from collections import defaultdict
import math
class MultinomialNB:
    def __init__(self, articles_per_tag):
        # Don't change the following two lines of code.
        self.articles_per_tag = articles_per_tag
        self.likelihood_per_word_per_tag = {}
        self.alpha = 1
        self.tags = self.articles_per_tag.keys()
        self.priors_per_tag = {}
        self.train()

    def train(self):
        tag_counts_map = {tag :len(self.articles_per_tag[tag]) for tag in self.tags}
        self.priors_per_tag = {tag: tag_counts_map[tag]/sum(tag_counts_map.values()) for tag in tag_counts_map.keys()}
        self.likelihood_per_word_per_tag = self.__get_word_likelihood_per_tag()


    def __get_word_likelihood_per_tag(self):
        word_frequencies_per_tag = defaultdict(lambda :{tag: 0 for tag in self.tags})
        total_word_frequencies_per_tag = defaultdict(int)
        for tag in self.tags:
            for article in  self.articles_per_tag[tag]:
                for word in article:
                    word_frequencies_per_tag += 1
                    total_word_frequencies_per_tag += 1
        # likelyhood
        ## if any unknown words not present then its probability =0.5
        word_likelihood_per_tag = defaultdict(lambda:{tag: 0.5 for tag in self.tags})
        for word,tag_map in word_frequencies_per_tag.items():
            for tag in tag_map:
                word_likelihood_per_tag[word][tag] = (word_frequencies_per_tag[word][tag]+1*self.alpha)/(total_word_frequencies_per_tag[tag] + 2*self.tag)
        return word_likelihood_per_tag



    def predict(self, article):
        ## all posterior = prior
        posterior_per_tag = {tag : math.log(prior) for tag, prior in self.prior_per_tag.items()}
        for word in article:
            for tag in self.tags:
                posterior_per_tag[tag] = posterior_per_tag[tag] + math.log(word_likelihood_per_tag[word][tag])
        return posterior_per_tag






```
